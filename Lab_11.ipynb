# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')

        for next_city in range(n):
            if mask & (1<<next_city) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1<<next_city))
                if new_cost < ans:
                    ans = new_cost

        dp[city][mask] = ans
        return ans

    return visit(0, 1)
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

print("Minimum travelling cost:", tsp(graph))
```

```python
import math
def get_user_graph_input():

    while True:
        try:
            n = int(input("Enter the number of cities: "))
            if n < 2:
                print("Please enter at least 2 cities.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter an integer.")

    print(f"Enter the {n}x{n} distance matrix. For distances between two cities, enter -1 if there is no direct path or if it's the same city. For the same city, it will be automatically set to 0.")
    graph = []
    for i in range(n):
        row = []
        while True:
            try:
                distances_str = input(f"Enter distances for city {i} (space-separated, {n} values): ").split()
                if len(distances_str) != n:
                    print(f"Please enter exactly {n} values.")
                    continue
                distances = [int(d) for d in distances_str]
                for j in range(n):
                    if i == j:
                        row.append(0)
                    elif distances[j] == -1:
                        row.append(float('inf'))
                    elif distances[j] < 0:
                        print("Distances cannot be negative (except -1 for no path). Please re-enter.")
                        break
                    else:
                        row.append(distances[j])
                else:
                    graph.append(row)
                    break
            except ValueError:
                print("Invalid input. Please enter integers for distances.")
    return graph

def tsp_with_path(graph):
    n = len(graph)
    dp = [[-1] * (1 << n) for _ in range(n)]

    path_tracker = [[-1] * (1 << n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1 << n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        min_cost = float('inf')
        next_optimal_city = -1

        for next_city in range(n):

            if not (mask & (1 << next_city)):
                cost = graph[city][next_city] + visit(next_city, mask | (1 << next_city))
                if cost < min_cost:
                    min_cost = cost
                    next_optimal_city = next_city

        dp[city][mask] = min_cost
        path_tracker[city][mask] = next_optimal_city
        return min_cost

    min_total_cost = visit(0, 1 << 0)

    path = []
    current_city = 0
    current_mask = 1 << 0

    while current_city != -1 and current_mask < (1 << n) - 1:
        path.append(current_city)
        next_city = path_tracker[current_city][current_mask]
        if next_city == -1:
            break
        current_city = next_city
        current_mask |= (1 << current_city)

    if min_total_cost != float('inf'):
        path.append(0)
    else:
        return float('inf'), []

    return min_total_cost, path

user_graph = get_user_graph_input()

if not user_graph:
    print("No graph provided. Exiting.")
else:
    print("\nCalculating TSP...")
    min_cost, optimal_path = tsp_with_path(user_graph)

    if min_cost == float('inf'):
        print("No valid path found for the given graph.")
    else:
        print(f"\nMinimum travelling cost: {min_cost}")
        path_str = " -> ".join(map(str, optimal_path))
        print(f"Optimal path of cities visited: {path_str}")

```
