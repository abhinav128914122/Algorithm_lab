# Warshall algorithm to find reachability between all pairs of vertices in a graph.
Problem Description:
Understand transitive closure of a graph. Use Python to compute reachability using matrix method.

#Theory
Warshall algorithm finds whether a path exists between every pair of vertices. It updates the adjacency matrix step by step. If there is a path from i to j through k, then the entry becomes 1.

## Algorithm
1. Represent graph using adjacency matrix.
2. For each intermediate vertex k
3. Check all pairs i and j
4. Update value if path exists through k
4. Repeat for all vertices


```python
def warshall(graph):

    n = len(graph)

    for k in range(n):
        for i in range(n):
            for j in range(n):
                graph[i][j] = graph[i][j] or (graph[i][k] and graph[k][j])

    return graph
```

```python
INF = float('inf')

def warshall(graph):
    n = len(graph)

    for k in range(n):
        for i in range(n):
            for j in range(n):
                graph[i][j] = graph[i][j] or (graph[i][k] and graph[k][j])

    return graph

def floyd(graph):
    n = len(graph)
    dist = [row[:] for row in graph]

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] != INF and dist[k][j] != INF:
                    dist[i][j] = min(dist[i][j], dist[i][k] + dist[k][j])

    return dist

n = 5
print("Enter adjacency matrix for 5 vertices (0 or 1):")

graph = []
for i in range(n):
    row = list(map(int, input(f"Row {i+1}: ").split()))
    graph.append(row)

warshall_graph = [row[:] for row in graph]

weighted_graph = []
for i in range(n):
    row = []
    for j in range(n):
        if i == j:
            row.append(0)
        elif graph[i][j] == 0:
            row.append(INF)
        else:
            row.append(1)
    weighted_graph.append(row)

warshall_result = warshall(warshall_graph)
floyd_result = floyd(weighted_graph)

print("\n Transitive Closure Matrix (Warshall):")
for row in warshall_result:
    print(row)

print("\n Path Existence (Readable Format):")
for i in range(n):
    for j in range(n):
        if warshall_result[i][j]:
            print(f"✔ Path exists from {i} → {j}")
        else:
            print(f"✘ No path from {i} → {j}")


print("\n Shortest Path Matrix (Floyd):")
for row in floyd_result:
    print(["∞" if x == INF else x for x in row])

print("\n Comparison:")
print(" Warshall tells WHETHER a path exists.")
print("Floyd tells SHORTEST DISTANCE between nodes.")
print("If Floyd[i][j] != ∞ → Path exists (same as Warshall).")
```

## Find shortest paths between all pairs of vertices using Floyd Warshall algorithm.

#Understand all pairs shortest path problem. Implement dynamic programming approach in Python.

* Given a weighted graph with positive or negative edge weights, find
shortest distances between every pair of vertices.

# Theory
Floyd Warshall algorithm computes shortest paths between all pairs of nodes. It updates the distance matrix by checking if a shorter path exists through an intermediate vertex. Works with negative weights but fails with negative cycles.

Algorithm
1. Represent graph using distance matrix.
2. Initialize matrix with given weights.
3. Set distance of node to itself as 0.
4. For each intermediate vertex k
5. Update distance[i][j] = min(distance[i][j], distance[i][k] + distance[k][j])
6. Repeat for all vertices

```python
def floyd_warshall(graph):

    n = len(graph)

    dist = []
    for i in range(n):
        dist.append(graph[i][:])

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]

    return dist

```

```python
INF = float('inf')

def floyd_warshall(graph):
    n = len(graph)

    dist = [row[:] for row in graph]

    nxt = [[None]*n for _ in range(n)]

    for i in range(n):
        for j in range(n):
            if graph[i][j] != INF and i != j:
                nxt[i][j] = j

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] != INF and dist[k][j] != INF:
                    if dist[i][k] + dist[k][j] < dist[i][j]:
                        dist[i][j] = dist[i][k] + dist[k][j]
                        nxt[i][j] = nxt[i][k]

    return dist, nxt

def has_negative_cycle(dist):
    n = len(dist)
    for i in range(n):
        if dist[i][i] < 0:
            return True
    return False

def get_path(u, v, nxt):
    if nxt[u][v] is None:
        return []

    path = [u]
    while u != v:
        u = nxt[u][v]
        path.append(u)
    return path

n = int(input("Enter number of vertices: "))

print("Enter adjacency matrix (use INF for no edge):")
graph = []

for i in range(n):
    row = input(f"Row {i+1}: ").split()
    temp = []
    for val in row:
        if val.upper() == "INF":
            temp.append(INF)
        else:
            temp.append(int(val))
    graph.append(temp)

dist, nxt = floyd_warshall(graph)

print("\n🔹 Shortest Distance Matrix:")
for row in dist:
    print(["INF" if x == INF else x for x in row])

if has_negative_cycle(dist):
    print("\nGraph contains a NEGATIVE CYCLE!")
else:
    print("\nNo negative cycle detected.")

u = int(input("\nEnter source vertex: ")) - 1
v = int(input("Enter destination vertex: ")) - 1

path = get_path(u, v, nxt)

if not path:
    print("No path exists.")
else:
    print(f"Shortest path from {u} to {v}: {' → '.join(map(str, path))}")
    print(f"Distance: {dist[u][v]}")

print("\n🔸 Comparison with Dijkstra Algorithm:")
print("- Floyd-Warshall computes shortest paths between ALL pairs.")
print("- Dijkstra computes shortest path from ONE source to all vertices.")
print("- Floyd works with negative weights (no negative cycles).")
print("- Dijkstra does NOT work with negative weights.")
print("- Time Complexity:")
print("  Floyd-Warshall: O(n^3)")
print("  Dijkstra: O((V + E) log V) using priority queue")
```
